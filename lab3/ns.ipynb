```python
import math

def comprovacio (list, q):
    sum = 0
    for i in list:
        sum += (q**-i)
    if sum <= 1: 
        return True
    else: 
        return False
```

```python
def dec2alf(aux, q, alf, l):    
    sol = ""
    while aux > 0:
        n = aux % q
        sol = str(alf[n])+ sol
        aux = math.floor(aux/q)
    while len(sol) < l:
        sol = str(alf[0]) + sol
    return sol
```

```python
def canonical_code (list, q, alf = []):
    list_copia = list.copy()
    if comprovacio(list, q):
        list_copia.sort()
        sol = []
        pre = 0
        aux = 0
        for i in list_copia:
            if pre != 0 and pre != i:
                aux = (q **(i - pre)) * aux
            pre = i
            sol.append(dec2alf(aux, q, alf, i))
            aux += 1
        sol_ordenat = []
        trobat = False
        for i in list:
            for j in sol:
                if i == len(j) and not trobat:
                    sol_ordenat.append(j)
                    sol.remove(j)
                    trobat = True
            trobat = False
        return sol_ordenat
    else: 
        return list(-1)
```

```python
def completar_nivel_grupo(p, k, code):
    aux = ""
    for i in range(0,len(p),k):
        for j in range(k):
            aux = aux+p[i+j]
        code.update({aux: code.get(aux)+1})
        aux = ""
    return code
```

```python
def huffman_code(src2, k):
    total = 0
    huf = {}
    src = src2.copy()
    for i in src:
        total += i[1]
        huf.update({i[0]: 0}) 
    while len(src) > 1:
        e_min = -1
        v_min = total
        e_min2 = -1
        v_min2 = total
        for i in range(len(src)):
            if src[i][1] < v_min:
                if v_min2 > v_min:
                    v_min2 = v_min
                    e_min2 = e_min
                v_min = src[i][1]
                e_min = i
            if src[i][1] < v_min2 and  e_min != i:
                v_min2 = src[i][1]
                e_min2 = i

        completar_nivel_grupo(src[e_min][0], k, huf)
        completar_nivel_grupo(src[e_min2][0], k, huf)

        src += [(src[e_min][0]+src[e_min2][0], src[e_min][1]+src[e_min2][1])]
        src.pop(e_min)
        if e_min < e_min2:
            src.pop(e_min2-1)
        else:
            src.pop(e_min2)
    can = canonical_code(list(huf.values()), 2, [0, 1])
    code = {}
    huf = list(huf)
    for i in range(len(huf)):
        code.update({huf[i]: can[i]})
    return can
```

```python
import math

def shannon_code(src):
    w = 0
    p = []
    for i in src:
        w += i[1]
    for i in src:
        p.append(math.ceil(-math.log2(i[1]/w)))
    return p
```

```python
def update_src(src):
    total = 0
    for i in src:
        total += i[1]
    src2 = []
    for i in src:
        for j in src:
            src2.append((i[0]+j[0], i[1]/total*j[1]/total))
    return src2
```

```python
import math

def H(src):
    total = 0
    for i in src:
        total += i[1]
    r = 0
    for i in src:
        r += i[1]/total * math.log2(total/i[1])
    return r 
```
