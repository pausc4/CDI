```python
def encoder_LZW(txt):
    A = sorted(list(set(txt)))
    t = 1
    tokens = []
    r = 0
    while r < len(txt):
        aux = min(t, len(txt)-r)
        for i in range (aux, 0, -1):
            palabra = txt[r:r+i]
            if palabra in A:
                pos = A.index(palabra)
                c = palabra + txt[min(r+i, len(txt)-1)]
                if t < len(c):
                    t = len(c)
                break
        r += len(c)-1
        tokens.append(pos)
        A.append(c)
    return tokens

def decoder_LZW(tokens, A):
    txt = A[tokens[0]]
    aux = A[tokens[0]]
    for i in range (1, len(tokens)):
        if tokens[i] >= len(A):
            c = aux + aux[0]
            A.append(c)
        else :
            c = A[tokens[i]]  
            A.append(aux+c[0])
        txt += c
        aux = c
        i += len(aux)
    return txt
```
